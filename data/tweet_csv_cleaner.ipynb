```python
import pandas as pd
import numpy as np
import copy
import re

df = pd.read_csv("SportsTweets.csv")
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.dropna(inplace=True)
```

```python
df.head(25)
```

```python
def expanded(phrase):
    # specific
    phrase = re.sub(r"won\'t", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

```python
for row in df.itertuples(index=True):

    # Strip the tweet for any redundant whitespaces
    df.loc[row.Index, "Tweet"] = df.loc[row.Index, "Tweet"].strip()
    

    # Change tweets to all lower case for easier language processing
    df.loc[row.Index, "Tweet"] = df.loc[row.Index, "Tweet"].lower()

    # Removes rt keyword for retweets
    df.loc[row.Index, "Tweet"] = re.sub("rt\s", "", df.loc[row.Index, "Tweet"])

    # expands contractions
    df.loc[row.Index, "Tweet"] = expanded(df.loc[row.Index, "Tweet"])

    # Removes @'s from tweets
    df.loc[row.Index, "Tweet"] = re.sub("@[A-Za-z0-9_:]+","", df.loc[row.Index, "Tweet"])

    # Removes #'s hashtags from tweets
    df.loc[row.Index, "Tweet"] = re.sub("#[A-Za-z0-9_-]*\s*","", df.loc[row.Index, "Tweet"])

    # Removes https:// and www. links from tweets
    df.loc[row.Index, "Tweet"] = re.sub("http\S+", "", df.loc[row.Index, "Tweet"])
    df.loc[row.Index, "Tweet"] = re.sub("www.\S+", "", df.loc[row.Index, "Tweet"])

    # Removes \r\n newline characters from the tweet
    df.loc[row.Index, "Tweet"] = df.loc[row.Index, "Tweet"].replace("\r\n", " ")

    # Remove parentheses, exclamation mark, question mark
    df.loc[row.Index, "Tweet"] = re.sub('[()!?]', ' ', df.loc[row.Index, "Tweet"])

    # Remove any none alphanumeric characters one more time
    df.loc[row.Index, "Tweet"] = re.sub("[^a-z0-9]"," ", df.loc[row.Index, "Tweet"])

    # Strips the tweet again just incase of any blank characters
    df.loc[row.Index, "Tweet"] = df.loc[row.Index, "Tweet"].strip()

    if len(df.loc[row.Index, "Tweet"]) == 0:
        df.loc[row.Index, "Tweet"] = np.NA



```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace = True)
```

```python
df = df.loc[df['Sport'].isin(['American football','Soccer', 'Basketball', 'Baseball', "Hockey"])]
```

```python
df.reset_index(drop=True, inplace=True)
```
